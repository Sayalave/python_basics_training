# Equivalent of `SELECT`, `FROM`, and `WHERE` in Pandas

```python
import pandas as pd
```

## Pandas

### Equivalent of `FROM`

- In SQL, `FROM` refers to the table that we are querying. 
- In Pandas, we have a variable that represents an object. So the equivalent of `FROM` is creating the the DataFrame object. 

```python
oscars = pd.read_csv("https://datahub.io/rufuspollock/oscars-nominees-and-winners/r/data.csv")
```

```python
oscars.head()
```

### Equivalent of `SELECT`

- In SQL, `SELECT` is an operation on the columns of the table. 
- The oscars dataset has four columns: 'year', 'category', 'winner', and 'entity'
- In order to select specific columns, we pass a list of column names to the DataFrame object

```python
oscars[['year', 'category']]
```

```python
oscars[['entity', 'year']]
```

### Equivalent of `WHERE`

- In SQL, `WHERE` acts on the rows of the table. 
- In pandas, we do this by selecting the columns and then the condition.

```python
# Return rows where the year was 2016
oscars[oscars.year == 2016]
```
